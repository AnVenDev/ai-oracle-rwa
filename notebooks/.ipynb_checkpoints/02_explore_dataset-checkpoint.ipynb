### Imports and CSV upload

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Carica il dataset
df = pd.read_csv("../data/property_dataset_mvp.csv")
df.head()
```

### Description statistics

```python
df.info()
df.describe(include='all')
```

### Categorical data analysis

```python
# Frequenze per location ed energy_class
print("Distribuzione delle città:")
print(df['location'].value_counts())

print("\nDistribuzione classi energetiche:")
print(df['energy_class'].value_counts())

# Grafico a barre
sns.countplot(data=df, x='energy_class', order=sorted(df['energy_class'].unique()))
plt.title("Distribuzione classi energetiche")
plt.show()
```

### Histograms and distributions

```python
# Dimensione e prezzo
sns.histplot(df['size_m2'], kde=True)
plt.title("Distribuzione delle dimensioni (m²)")
plt.show()

sns.histplot(df['valuation_k'], kde=True)
plt.title("Distribuzione del valore stimato (in migliaia di €)")
plt.show()
```

### Heatmap of numeric correlations

```python
# Seleziona solo colonne numeriche
num_df = df.select_dtypes(include=[np.number])

# Calcola matrice di correlazione
corr = num_df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlazione tra variabili numeriche")
plt.show()
```

### Scatterplot: size vs valuation

```python
sns.scatterplot(data=df, x="size_m2", y="valuation_k", hue="energy_class")
plt.title("Dimensione vs Valore stimato")
plt.show()
```

### Boxplot value by city

```python
sns.boxplot(data=df, x="location", y="valuation_k")
plt.title("Valore stimato per città")
plt.xticks(rotation=45)
plt.show()
```
