### Imports and CSV upload

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Carica il dataset
df = pd.read_csv("../data/property_dataset_mvp_multi_rwa.csv")
df.head()
df[['asset_type','condition_score','risk_score','last_verified_ts']].head()
```

### Description statistics

```python
df.info()
df.describe(include='all')
```

### Key fields

```python
print("Shape:", df.shape)
print("\nMissing values per column:")
print(df.isnull().sum())

print("\nAsset types distribution:")
print(df['asset_type'].value_counts())
```

### Categorical data analysis

```python
print("Distribuzione delle città:")
print(df['location'].value_counts(), "\n")

print("Distribuzione classi energetiche:")
print(df['energy_class'].value_counts(), "\n")

# Tabella percentuali energy class
energy_order = ['A','B','C','D','E','F','G']
energy_counts = df['energy_class'].value_counts().reindex(energy_order)
energy_pct = (energy_counts / len(df) * 100).round(1)
display(pd.DataFrame({'count': energy_counts, 'pct%': energy_pct}))

# Countplot energy class
sns.countplot(data=df, x='energy_class', order=energy_order)
plt.title("Distribuzione classi energetiche")
plt.show()

# Countplot location (orizzontale)
sns.countplot(data=df, y='location', order=df['location'].value_counts().index)
plt.title("Distribuzione location")
plt.show()
```

### Key distributions (size and value)

```python
fig, axes = plt.subplots(1, 2, figsize=(16,6))
sns.histplot(df['size_m2'], kde=True, ax=axes[0])
axes[0].set_title("Distribuzione dimensioni (m²)")
sns.histplot(df['valuation_k'], kde=True, ax=axes[1])
axes[1].set_title("Distribuzione valore stimato (k€)")
plt.show()
```

```python
# Dimensione e prezzo
sns.histplot(df['size_m2'], kde=True)
plt.title("Distribuzione delle dimensioni (m²)")
plt.show()

sns.histplot(df['valuation_k'], kde=True)
plt.title("Distribuzione del valore stimato (in migliaia di €)")
plt.show()
```

### Condition and risk

```python
fig, axes = plt.subplots(1, 2, figsize=(16,6))
sns.histplot(df['condition_score'], kde=True, ax=axes[0], color="green")
axes[0].set_title("Distribution of Condition Score")
sns.histplot(df['risk_score'], kde=True, ax=axes[1], color="red")
axes[1].set_title("Distribution of Risk Score")
plt.show()

# Scatter inverso condizione vs rischio
sns.scatterplot(data=df, x='condition_score', y='risk_score', hue='energy_class', palette='viridis')
plt.title("Condition score vs Risk score (should be roughly inverse)")
plt.show()

# Boxplot condition per energy class
sns.boxplot(data=df, x='energy_class', y='condition_score', order=energy_order)
plt.title("Condition Score per Energy Class")
plt.show()
```

### Relations

```python
# Correlazione numeriche (includi nuove metriche)
num_df = df.select_dtypes(include=[np.number])
corr = num_df.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))  # maschera triangolo superiore
plt.figure(figsize=(14,10))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f",
            cmap='coolwarm', center=0, square=True,
            cbar_kws={"shrink": .8})
plt.title("Correlation Heatmap (lower triangle)")
plt.show()
```

```python
# Scatter dimensione vs valore
sns.scatterplot(data=df, x="size_m2", y="valuation_k", hue="energy_class", palette="deep")
plt.title("Dimensione vs Valore stimato (colorato per energy_class)")
plt.show()
```

```python
# Valore stimato per energy_class
avg_val_by_energy = df.groupby('energy_class')['valuation_k'].mean().reindex(energy_order)
sns.barplot(x=avg_val_by_energy.index, y=avg_val_by_energy.values)
plt.title("Valore medio (k€) per classe energetica")
plt.ylabel("valuation_k")
plt.show()
```

```python
# Boxplot valore per città
sns.boxplot(data=df, x="location", y="valuation_k", order=df['location'].value_counts().index)
plt.title("Valore stimato per città")
plt.xticks(rotation=45)
plt.show()
```
